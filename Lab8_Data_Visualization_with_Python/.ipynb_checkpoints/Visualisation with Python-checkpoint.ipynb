```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading a dataset:
healthcare_dataframe = pd.read_csv('healthcare_dataset.csv')
```

```python
# Reading first 10 rows of data
healthcare_dataframe.head(10)
```

```python
# Retrieving information about fields
healthcare_dataframe.info()
```

```python
# Casting Date of Admission type to date
healthcare_dataframe['Date of Admission'] = pd.to_datetime(healthcare_dataframe['Date of Admission'])

# Creating Admission Year field
healthcare_dataframe['Admission Year'] = healthcare_dataframe['Date of Admission'].dt.year
```

```python
# Obtaining preliminary understanding of numeric fields
healthcare_dataframe.describe()
```

```python
# Retrieving details about medical conditions 
healthcare_dataframe["Medical Condition"].value_counts()
```

```python
# Medical Condition
# Arthritis       9308
# Diabetes        9304
# Hypertension    9245
# Obesity         9231
# Cancer          9227
# Asthma          9185

# Medical conditions per year
medical_condition_color = {
    "Arthritis": "#e41a1c",     # Red
    "Diabetes": "#377eb8",      # Blue
    "Hypertension": "#4daf4a",  # Green
    "Obesity": "#984ea3",       # Purple
    "Cancer": "#ff7f00",        # Orange
    "Asthma": "#a65628",        # Brown
}

for condition, condition_color in medical_condition_color.items():
    # creating a data frame for a specific value for 'Medical Condition' column:
    medical_conditions_dataframe = healthcare_dataframe[healthcare_dataframe["Medical Condition"] == condition]
    # parameter c - color
    plt.scatter(medical_conditions_dataframe["Age"], medical_conditions_dataframe["Medical Condition"], c = condition_color, label=condition)

# plt.grid(True)
# plt.title("Iris Petal Length vs Width")
# plt.xlabel("Petal Length (cm)")
# plt.ylabel("Petal Width (cm)")
# plt.legend()
```
